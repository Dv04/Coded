  # Control Statements

### If syntax:

Statements x
if ( condition ):
    Statements block
Statements y

### If Definition:

After the first x statements are run, the system checks the condition of the IF, if it is true, the block of statements is run, otherwise the system skips the block and continues with the y statements.

### If Example:

x = 1
if ( x == 1 ):
    print("x is 1")
print("x may be 1")


### Elif syntax:

Statements x
if ( condition ):
    Statements block
elif ( condition ):
    Statements block
elif ( condition ):
    Statements block

### Elif definition:

After the first x statements are run, the system checks the condition of the IF, if it is true, the block of statements is run, otherwise the system checks the condition of the first ELIF, if it is true, the block of statements is run, otherwise the system checks the condition of the second ELIF, if it is true, the block of statements is run, otherwise the system skips the block and continues with the y statements.

### Elif example:

x = 1
if ( x == 1 ):
    print("x is 1")
elif ( x == 2 ):
    print("x is 2")
elif ( x == 3 ):
    print("x is 3")
print("x may be 1, 2 or 3")


### Else syntax:

Statements x
if ( condition ):
    Statements block
else:
    Statements block
Statements y

### Else definition:

After the first x statements are run, the system checks the condition of the IF, if it is true, the block of statements is run, otherwise the system runs the block of statements of the ELSE, and then continues with the y statements.

### Else example:

x = 1
if ( x == 1 ):
    print("x is 1")
else:
    print("x is not 1")
print("x may be 1 or not 1")
