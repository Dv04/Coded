# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array(np.random.randint(0, 100, 50))
y = np.array(np.random.randint(0, 100, 50))
plt.plot(x,y, color = "black", linewidth = 1.0, linestyle = "-")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Matplotlib Trial")
plt.autoscale(1,"x")
plt.arrow(0,0,100,100, color = "black", linewidth = 1.0, linestyle = "--")
plt.show()
```

```python
plt.scatter(x,y, color = "black")
plt.show()
```

```python
plt.bar(x,y, color = "black")
plt.show()
```

```python
plt.hist(x,range=(0,100), bins=10, color = "red", label = "X",rwidth=0.5,orientation="vertical")
plt.xticks(np.arange(0,100,1))
plt.show()
```

```python
plt.pie(y,labels=x,shadow=False,radius=5.0,startangle=90,autopct="%2.3f%%",textprops={"fontsize":10})
```
