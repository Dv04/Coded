```python
import numpy as np
import matplotlib.pyplot as plt

# Example data
x = np.random.rand(50)
y = 2*x + np.random.normal(0, 0.1, (50))

# Calculating coefficients m (slope) and c (intercept)
N = len(x)
sum_x = np.sum(x)
sum_y = np.sum(y)
sum_xy = np.sum(x*y)
sum_xx = np.sum(x*x)

# Slope (m)
m = (N*sum_xy - sum_x*sum_y) / (N*sum_xx - sum_x**2)

# Intercept (c)
c = (sum_y - m*sum_x) / N

print("Slope (m):", m)
print("Intercept (c):", c)

# Plotting the regression line
plt.scatter(x, y, color='red', label='Data points')
plt.plot(x, m*x + c, label='Regression Line')
plt.xlabel('Independent variable (x)')
plt.ylabel('Dependent variable (y)')
plt.title('Least Squares Regression Line')
plt.legend()
plt.show()
```
